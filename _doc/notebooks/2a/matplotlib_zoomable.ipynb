# Graphe zoomable avec Matplotlib

Cela ne fonctionne que dans un notebook et avec la version 1.4.2+ de matploblib. Le cas classique en style [ggplot](http://matplotlib.org/users/style_sheets.html) parce que c'est plus joli :

```
from matplotlib import pyplot as plt
plt.style.use('ggplot')
```

On essaye avec un graphe quelconque :

```
import random
x = [ random.random() for i in range(0,100) ]
y = [ 2*_ + random.random()/2 for _ in x ]
ax = plt.subplot('111')
ax.plot (x,y,"o")
```

On passe à un affichage en javascript (il ne vaut mieux changer sans arrêt de cela crée quelques problèmes d'affichages).

```
plt.switch_backend('nbagg')
```
